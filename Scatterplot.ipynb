```python
name = data['Rating']
price = data['Reviews']
 
# Figure Size
fig = plt.figure(figsize =(16, 7))
plt.xlabel("Rating")
plt.ylabel("Reviews")
 
# Horizontal Bar Plot
plt.scatter(name[10:1082], price[10:1082],c = 'r',marker = '*' ,alpha = 0.5,linewidth = 10,edgecolor = 'g')
 
# Show Plot
plt.show()
```

```python
name = data['Rating']
price = data['Installs']
 
# Figure Size
fig = plt.figure(figsize =(16, 7))
plt.xlabel("Rating")
plt.ylabel("Installs & Rating")
 
# Horizontal Bar Plot
plt.scatter(name[10:100], price[10:100],c = 'r',marker = '*' ,linewidth = 10,edgecolor = 'g')
plt.scatter(name[10:100], price[10:100],c = 'g',marker = 'o' ,linewidth = 10,edgecolor = 'g')
 
# Show Plot
plt.show()
```
